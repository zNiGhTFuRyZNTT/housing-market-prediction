```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the original dataset
original_file_path = 'MELBOURNE_HOUSE_PRICES_LESS.csv'  # Change to your original dataset path
original_data = pd.read_csv(original_file_path)

# Load the cleaned dataset
cleaned_file_path = 'MELBOURNE_HOUSE_PRICES_CLEANED.csv'  # Change to your cleaned dataset path
cleaned_data = pd.read_csv(cleaned_file_path)

# Select only numeric columns for correlation calculation
original_numeric = original_data.select_dtypes(include=['float64', 'int64'])
cleaned_numeric = cleaned_data.select_dtypes(include=['float64', 'int64'])

# Create a heatmap for correlation matrices
plt.figure(figsize=(12, 6))

# Original dataset heatmap
plt.subplot(1, 2, 1)
sns.heatmap(original_numeric.corr(), annot=True, cmap='coolwarm', cbar=True)
plt.title("Original Dataset - Correlation Matrix")

# Cleaned dataset heatmap
plt.subplot(1, 2, 2)
sns.heatmap(cleaned_numeric.corr(), annot=True, cmap='coolwarm', cbar=True)
plt.title("Cleaned Dataset - Correlation Matrix")

# Save and display the heatmap comparison
plt.tight_layout()
plt.savefig('heatmap_comparison.png')
plt.show()

# Create a box plot for Price distribution
plt.figure(figsize=(12, 6))

# Original dataset box plot
plt.subplot(1, 2, 1)
sns.boxplot(y=original_data['Price'])
plt.title("Original Dataset - Price Distribution")

# Cleaned dataset box plot
plt.subplot(1, 2, 2)
sns.boxplot(y=cleaned_data['Price'])
plt.title("Cleaned Dataset - Price Distribution")

# Save and display the box plot comparison
plt.tight_layout()
plt.savefig('boxplot_comparison.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the original dataset
original_file_path = 'Melbourne_housing_FULL.csv'  # Change to your original dataset path
original_data = pd.read_csv(original_file_path)

# Load the cleaned dataset
cleaned_file_path = 'Melbourne_housing_FULL_CLEANED.csv'  # Change to your cleaned dataset path
cleaned_data = pd.read_csv(cleaned_file_path)

# Choose a valid categorical column for comparison
categorical_column = 'Suburb'  # Replace with a valid column name in your dataset

# Create a bar plot for the chosen categorical column count comparison
plt.figure(figsize=(14, 6))

# Original dataset bar plot
plt.subplot(1, 2, 1)
sns.countplot(x=categorical_column, data=original_data, palette='viridis')
plt.xticks(rotation=45)
plt.title(f"Original Dataset - {categorical_column} Count")

# Cleaned dataset bar plot
plt.subplot(1, 2, 2)
sns.countplot(x=categorical_column, data=cleaned_data, palette='viridis')
plt.xticks(rotation=45)
plt.title(f"Cleaned Dataset - {categorical_column} Count")

# Save and display the bar plot comparison
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the original dataset
original_file_path = 'combined_housing_data.csv'  # Change to your original dataset path
original_data = pd.read_csv(original_file_path)

# Load the cleaned dataset
cleaned_file_path = 'combined_housing_data_CLEANED.csv'  # Change to your cleaned dataset path
cleaned_data = pd.read_csv(cleaned_file_path)

# ------------------- Pie Chart -------------------
# Create a pie chart for Type distribution in the original and cleaned datasets
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
original_data['Type'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Original Dataset - Property Type Distribution')
plt.ylabel('')  # Remove y-label for aesthetics

plt.subplot(1, 2, 2)
cleaned_data['Type'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Cleaned Dataset - Property Type Distribution')
plt.ylabel('')  # Remove y-label for aesthetics

plt.tight_layout()
plt.show()

# ------------------- Histogram -------------------
# Create a histogram for Price distribution
plt.figure(figsize=(14, 7))

# Original dataset histogram
plt.subplot(1, 2, 1)
plt.hist(original_data['Price'].dropna(), bins=30, color='skyblue', alpha=0.7)
plt.title('Original Dataset - Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Cleaned dataset histogram
plt.subplot(1, 2, 2)
plt.hist(cleaned_data['Price'].dropna(), bins=30, color='salmon', alpha=0.7)
plt.title('Cleaned Dataset - Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# ------------------- Heatmap -------------------
# Select only numeric columns for correlation calculation
original_numeric = original_data.select_dtypes(include=['float64', 'int64'])
cleaned_numeric = cleaned_data.select_dtypes(include=['float64', 'int64'])

# Create a heatmap for correlation matrices
plt.figure(figsize=(14, 7))

# Original dataset heatmap
plt.subplot(1, 2, 1)
sns.heatmap(original_numeric.corr(), annot=True, cmap='coolwarm', cbar=True)
plt.title("Original Dataset - Correlation Matrix")

# Cleaned dataset heatmap
plt.subplot(1, 2, 2)
sns.heatmap(cleaned_numeric.corr(), annot=True, cmap='coolwarm', cbar=True)
plt.title("Cleaned Dataset - Correlation Matrix")

plt.tight_layout()
plt.show()
```
